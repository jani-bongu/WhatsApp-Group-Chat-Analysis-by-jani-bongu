```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with Bitter bananas🔥.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
date=[]
times=[]
for i in dates:
    date.append(i.split(", ")[0])
    times.append(i.split(", ")[1])
```

```python
times
```

```python
time=[]
for i in times:
    time.append(i.split(" - ")[0])
```

```python
df=pd.DataFrame({'user_message':messages,
                 'date':date,
                'time':time})
```

```python
df.head(5)
```

```python
# split into user and message 
```

```python
user=[]
message=[]
for i in df['user_message']:
    x=re.split("([\w\W]+?):\s",i)
    if x[1:]:
        user.append(x[1])
        message.append(x[2])
    else:
        user.append('Group Notification')
        message.append(x[0])
```

```python
df['user']=user
df['message']=message
df.drop(columns=['user_message'],inplace=True)
```

```python
df.head(20)
```

```python
df['date']=pd.to_datetime(df['date'],format='%d/%m/%y')
```

```python
df['time']=pd.to_datetime(df['time'],format='%I:%M %p').dt.strftime('%H:%M')
```
